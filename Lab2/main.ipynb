```python
from nltk.corpus import brown
news_text = brown.words(categories='news')
# print(news_text)
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m])

print("=" * 10)

science_fiction = brown.words(categories='science_fiction')
fdist_1 = nltk.FreqDist([w.lower() for w in science_fiction])
modals = ['what', 'where', 'who', 'why', 'whom', 'when', 'which', 'whose', 'whatever']
for m in modals:
    print(m + ':', fdist_1[m])
```

```python
cfd = nltk.ConditionalFreqDist(
    (genre, word) for genre in brown.categories() for word in brown.words(categories=genre)
)

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
cfd.tabulate(conditions = genres, samples = modals)
```

```python
max(modals, key=lambda w: cfd['news'][w]), max(modals, key=lambda w: cfd['romance'][w])
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return round(len(content) / len(text), 2)

content_fraction(nltk.corpus.reuters.words())
```

```python
mytexts = nltk.TextCollection([text1, text2, text3])
```

TF (Term Frequency): The goal is to measure how frequently a term occurs in a specific document to determine its local importance.

IDF (Inverse Document Frequency): The goal is to measure how rare a term is across the entire corpus to penalize common "stop words" and reward unique, informative words.

TF-IDF: The goal is to combine local frequency and global rarity to assign a weight that represents how relevant a specific word is to a specific document within a collection.

```python
mytexts.tf('is', text1), mytexts.tf('is', text2), mytexts.tf('is', text3)
```

```python
mytexts.idf('companionableness')
```

```python
mytexts.tf_idf('companionableness', text2)
```
